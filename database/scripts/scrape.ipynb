```python
import yfinance as yf
import pandas as pd
```

```python
tickers = ['AAPL', 'MSFT', 'GOOGL']
portfolio_df = pd.concat([yf.download(ticker, group_by='Ticker', period='2y', interval='1d').assign(ticker=ticker) for ticker in tickers], ignore_index=False)
portfolio_df
```

```python
import os
cwd = os.getcwd()
path = cwd + "/data_snapshot.csv"
portfolio_df.to_csv(path)
```
